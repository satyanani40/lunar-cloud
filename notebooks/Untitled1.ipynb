```python
class WordCloud(dict):
    def __init__(self):
        pass
    def __setitem__(self, key, val):
        if key in self:
            l=self[key]
            l.append(val)
            super(WordCloud, self).__setitem__(key, l)
        else:
            super(WordCloud, self).__setitem__(key, [val,])
        

x=WordCloud()

```

```python
from nltk.corpus import wordnet
```

```python
a=wordnet.synsets('fall')
b=wordnet.synsets('spill')
l=[]
for i in a:
    for j in b:
        similarity = i.wup_similarity(j)
        if similarity:
            l.append(similarity)
l.sort()
print l 
```
